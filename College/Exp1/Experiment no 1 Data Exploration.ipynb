````python
#print the contigency table between two categorical values black and marr to find chisquare value.
var2 = pd.crosstab(index=var['black'],columns=var['marr'],margins=True)
print(var2)
````

````python
#We will find the expected value using the chi2_contigency method of stats
stat, p, dof, expected = chi2_contingency(var2)
print("Statistics : ",stat)
print("Degree of Freedom : ",dof)
print("Expected value : " , expected)
````

# Find the Chi squared Value

````python
#Find the Chisquare value usng observed and expected by using chisquare method of scipy stat
X_2 = chisquare(var2 , expected)
X_2
````

# Finding CorrelationBy converting the age (numerical Value into Categorical Value)
The value were divided into 3 ranges according to their ages (Youth , Middle-Aged , Senior)
We used scikit learn LableEncoder and fit transform method to convert the numerical value into categorical value


````python
dataset = pd.read_csv('age_cat.csv')
#Dataset was too large so we reduced the data 
````

````python
X = dataset.values
````

````python
dataset
````

The dataset was divided into ranges(Youth, Middle, Senior-Citizen)

````python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#sklearn a scikit library use for preprocessing in Python 
````

````python
labelencoder_X = LabelEncoder()
#Calling the labelencoder function 
````

````python
X[:,0]= labelencoder_X.fit_transform(X[:,0])
````

````python
onehotencoder = OneHotEncoder(categorical_features = [0])
#Onehotencoder method which is use to convert categorical age to binary data 
#with middle,senior,youth representing column 0,1,2 respectively
````

````python
X = onehotencoder.fit_transform(X).toarray()
#converting into array of (1,0) binary data
````

````python
X.astype(int)
X = pd.DataFrame(X)
X.astype(int)
X
````

The Age is converted into category 

````python
X.corr(method = "kendall")
````
